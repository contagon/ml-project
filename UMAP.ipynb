```python
metrics   = ["cosine", "euclidean"]
neighbors = [5, 15, 30, 60]

fig, ax = plt.subplots(len(metrics), len(neighbors), figsize=(15,7))
fig.suptitle("Recipe TF-IDF - UMAP (n_neighbors and metric)")
for i in range(len(metrics)):
    for j in range(len(neighbors)):
        print(i,j)
        #perform and plot UMAP
        umapped = umap.UMAP(verbose=False, metric=metrics[i], n_neighbors=neighbors[j]).fit_transform(Xhat[idx])
        #iterate through all options
        sns.scatterplot(umapped[:,0], umapped[:,1], ax=ax[i,j])
        if i == 0:
            ax[i,j].set_title(f"{neighbors[j]}")
        if j == 0:
            ax[i,j].set_ylabel(f"{metrics[i]}")

plt.savefig("figs/recipeUMAP_tfidf.pdf", bbox_inches="tight")
plt.show()
```
